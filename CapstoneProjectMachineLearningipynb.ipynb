# Task
Generate visualizations to explore the distribution of features and their relationship with the target variable in the provided dataset. Summarize the key findings from the visualizations.

## Visualize the distribution of categorical features.

### Subtask:
Visualize the distribution of categorical features.


**Reasoning**:
Generate bar plots for 'Department' and 'Job_Title' columns to visualize their distributions.



```python
plt.figure(figsize=(10, 6))
df['Department'].value_counts().plot(kind='bar')
plt.title('Distribution of Employees by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
df['Job_Title'].value_counts().plot(kind='bar')
plt.title('Distribution of Employees by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Visualize the distribution of numerical features.

### Subtask:
Visualize the distribution of numerical features.


**Reasoning**:
Create histograms for the specified numerical columns to visualize their distributions.



```python
numerical_features = ['Age', 'Years_at_Company', 'Satisfaction_Level', 'Average_Monthly_Hours', 'Salary']

for col in numerical_features:
    plt.figure(figsize=(8, 6))
    df[col].hist(bins=20)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

## Visualize the relationship between features and the target variable.

### Subtask:
Visualize the relationship between features and the target variable.


**Reasoning**:
Create bar plots to visualize the relationship between categorical features ('Gender', 'Department', 'Job_Title', 'Promotion_Last_5Years') and the target variable 'Attrition'.



```python
import matplotlib.pyplot as plt
import seaborn as sns
categorical_features = ['Gender', 'Department', 'Job_Title', 'Promotion_Last_5Years']

for col in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=col, hue='Attrition')
    plt.title(f'Attrition by {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

**Reasoning**:
Import the seaborn library to use the `countplot` function and regenerate the bar plots for the relationship between categorical features and 'Attrition'.



```python
import seaborn as sns

categorical_features = ['Gender', 'Department', 'Job_Title', 'Promotion_Last_5Years']

for col in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=col, hue='Attrition')
    plt.title(f'Attrition by {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

**Reasoning**:
Create box plots to visualize the relationship between numerical features and the target variable 'Attrition'.



```python
numerical_features = ['Age', 'Years_at_Company', 'Satisfaction_Level', 'Average_Monthly_Hours', 'Salary']

for col in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, x='Attrition', y=col)
    plt.title(f'{col} vs Attrition')
    plt.xlabel('Attrition (0: No, 1: Yes)')
    plt.ylabel(col)
    plt.show()
```
