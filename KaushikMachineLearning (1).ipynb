```python
def is_prime(num):
  """Checks if a number is prime using a loop."""
  if num <= 1:
    return False  # Numbers less than or equal to 1 are not prime
  for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
      return False  # Found a divisor, so not prime
  return True  # No divisors found, so prime

# Example usage:
number_to_check = 17
if is_prime(number_to_check):
  print(f"{number_to_check} is a prime number.")
else:
  print(f"{number_to_check} is not a prime number.")

number_to_check = 12
if is_prime(number_to_check):
  print(f"{number_to_check} is a prime number.")
else:
  print(f"{number_to_check} is not a prime number.")
```

```python
def recursive_factorial(n):
  """Calculates the factorial of a non-negative integer using recursion."""
  if n == 0:
    return 1
  else:
    return n * recursive_factorial(n - 1)

def recursive_fibonacci(n):
  """Generates the nth Fibonacci number using recursion."""
  if n <= 1:
    return n
  else:
    return recursive_fibonacci(n - 1) + recursive_fibonacci(n - 2)

# Example usage for factorial:
num_factorial = 5
print(f"The factorial of {num_factorial} is {recursive_factorial(num_factorial)}")

# Example usage for Fibonacci:
num_fibonacci = 7
print(f"The {num_fibonacci}th Fibonacci number is {recursive_fibonacci(num_fibonacci)}")
```

```python
def is_palindrome(input_string):
  """Checks if a string is a palindrome."""
  # Remove spaces and convert to lowercase for case-insensitive check
  cleaned_string = input_string.replace(" ", "").lower()
  return cleaned_string == cleaned_string[::-1]

# Example usage:
string1 = "racecar"
if is_palindrome(string1):
  print(f"'{string1}' is a palindrome.")
else:
  print(f"'{string1}' is not a palindrome.")

string2 = "hello world"
if is_palindrome(string2):
  print(f"'{string2}' is a palindrome.")
else:
  print(f"'{string2}' is not a palindrome.")

string3 = "A man a plan a canal Panama"
if is_palindrome(string3):
  print(f"'{string3}' is a palindrome.")
else:
  print(f"'{string3}' is not a palindrome.")
```

```python
import numpy as np

def calculate_entropy(data):
  """Calculates the entropy of the input data."""
  unique_values, counts = np.unique(data, return_counts=True)
  probabilities = counts / len(data)
  entropy = -np.sum(probabilities * np.log2(probabilities + 1e-9))  # Add a small epsilon to avoid log(0)
  return entropy
```

# Task
Create a Python program that predicts customer purchases based on demographics and spending habits using both supervised and unsupervised learning techniques. The program should generate synthetic data, perform customer segmentation using clustering, and then train a classification model to predict purchases. Finally, the program should evaluate the classification model and interpret the results.

## Generate synthetic data

### Subtask:
Create a synthetic dataset that includes demographic information, spending habits, and a purchase outcome (binary: purchased/not purchased).


**Reasoning**:
The subtask is to create a synthetic dataset with demographic information, spending habits, and a purchase outcome. I will generate the data using numpy and then create a pandas DataFrame.



```python
import pandas as pd
import numpy as np

num_samples = 1000

# Generate synthetic data
age = np.random.randint(18, 70, size=num_samples)
income = np.random.uniform(20000, 100000, size=num_samples)
gender = np.random.choice(['Male', 'Female'], size=num_samples)
spending_score = np.random.randint(1, 100, size=num_samples)

# Generate 'Purchased' based on a combination of features
# Higher probability for older, higher income, higher spending score
purchase_probability = 1 / (1 + np.exp(-(0.05 * age + 0.00005 * income + 0.03 * spending_score - 8))) # Sigmoid function
purchased = (np.random.rand(num_samples) < purchase_probability).astype(int)

# Create DataFrame
df = pd.DataFrame({
    'Age': age,
    'Income': income,
    'Gender': gender,
    'Spending_Score': spending_score,
    'Purchased': purchased
})

# Display the first few rows
display(df.head())
```

## Unsupervised learning (clustering)

### Subtask:
Apply an unsupervised learning algorithm (e.g., K-Means) to the demographic and spending habit features to identify customer segments.


**Reasoning**:
Select the numerical features for clustering, import KMeans, instantiate the model, fit it to the data, and get the cluster labels.



```python
# Select numerical features for clustering
X_clustering = df[['Age', 'Income', 'Spending_Score']]

# Import KMeans
from sklearn.cluster import KMeans

# Instantiate KMeans model
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)

# Fit the model to the data
kmeans.fit(X_clustering)

# Get cluster labels
df['Cluster'] = kmeans.labels_

# Display the first few rows with the new cluster column
display(df.head())
```

## Feature engineering

### Subtask:
Add the cluster labels as a new feature to the dataset.


**Reasoning**:
The previous step already added the cluster labels to the DataFrame `df`. I will display the head of the DataFrame to verify the addition of the 'Cluster' column.



```python
display(df.head())
```

## Supervised learning (classification)

### Subtask:
Use the original features (demographics, spending habits) along with the new cluster feature to train a supervised learning model (e.g., Logistic Regression, RandomForest) to predict the purchase outcome.


**Reasoning**:
Select features and target, handle categorical features using one-hot encoding, split data into training and testing sets, and train a Logistic Regression model.



```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Select features and target
X = df[['Age', 'Income', 'Gender', 'Spending_Score', 'Cluster']]
y = df['Purchased']

# Identify categorical and numerical features
categorical_features = ['Gender']
numerical_features = ['Age', 'Income', 'Spending_Score', 'Cluster']

# Create a column transformer to handle one-hot encoding for the categorical feature
preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='passthrough' # Keep numerical features as they are
)

# Create a pipeline with preprocessing and the classifier
model = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', LogisticRegression(random_state=42))])

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the logistic regression model
model.fit(X_train, y_train)
```

## Evaluate the supervised model

### Subtask:
Evaluate the performance of the supervised model using appropriate metrics (e.g., accuracy, precision, recall, F1-score).


**Reasoning**:
I will make predictions on the test set, calculate accuracy, generate a classification report, and generate a confusion matrix to evaluate the model's performance. Then I will print these metrics.



```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
matrix = confusion_matrix(y_test, y_pred)

# Print evaluation metrics
print(f"Accuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(report)
print("\nConfusion Matrix:")
print(matrix)
```

Analyze the results of both the clustering and classification steps to understand the relationship between customer segments, demographics, spending habits, and purchase behavior.

To analyze the customer segments, I will group the dataframe by the 'Cluster' column and calculate the mean of the relevant features for each cluster. This will allow me to understand the characteristics of each customer segment. Then, I will print the results and provide a brief interpretation of the findings.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler


log_grid = {
    'log__C': [0.01, 0.1, 1, 10],
    'log__penalty': ['l2'],
    'log__solver': ['lbfgs', 'liblinear']
}

log_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('log', LogisticRegression(max_iter=1000))
])

grid_log = GridSearchCV(log_pipe, log_grid, cv=5, scoring='accuracy')
grid_log.fit(X_train, y_train)

print("Best Logistic Regression:", grid_log.best_params_)
print("Best Accuracy:", grid_log.best_score_) # Corrected to access the attribute
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Get the best model from the grid search
best_log_model = grid_log.best_estimator_

# Make predictions on the test set
y_pred_log = best_log_model.predict(X_test)

# Evaluate the model
accuracy_log = accuracy_score(y_test, y_pred_log)
report_log = classification_report(y_test, y_pred_log)
matrix_log = confusion_matrix(y_test, y_pred_log)

# Print evaluation metrics
print(f"Best Logistic Regression Model Accuracy: {accuracy_log:.4f}")
print("\nBest Logistic Regression Model Classification Report:")
print(report_log)
print("\nBest Logistic Regression Model Confusion Matrix:")
print(matrix_log)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler


rf_grid = {
    'rf__n_estimators': [100, 200, 300],
    'rf__max_depth': [None, 10, 20],
    'rf__min_samples_split': [2, 5, 10]
}

rf_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('rf', RandomForestClassifier(random_state=42)) # Added missing parenthesis
])

grid_rf = GridSearchCV(rf_pipe, rf_grid, cv=5, scoring='accuracy')
grid_rf.fit(X_train, y_train)

print("Best RF:", grid_rf.best_params_)
print("Best RF Score:", grid_rf.best_score_)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


X_cluster = df_encoded.drop(['Response'], axis=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

inertia = []
k_range = range(1, 11)

for k in k_range:
  kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # Added n_init
  kmeans.fit(X_scaled)
  inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o', linestyle='-', color='purple')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```
